```python
# Import cleaned accident data from csv created at end of accident cleaning notebook
df = pd.read_csv('data/Accidents/Cleaned_accidents.csv')
print('Records:', df.shape[0], '\nColumns:', df.shape[1])
# Dropping unnamed column
df.drop(df.columns[0],axis=1,inplace=True)
df.head()
```

```python
# #unique values
df.nunique().sort_values(ascending=False)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Month'] = df ['Date'].apply(lambda time: time.month)
```

```python
#creating a weekend feature that includes Friday-Sunday
df['Weekend']= np.where(df['Day_of_Week'].isin(['Friday', 'Saturday', 'Sunday']), 1, 0)
```

```python
#create time of day feature with Morning Rush, Day, Noon Rush, Afternoon, After Work Rush, Night

#time of day dictionary
timeofdaygroups = {1: "Morning Rush (6-10)",
                   2: "Day (10-12)",
                   3: "Lunch Rush (12-14)",
                   4: "Afternoon (14-16)",
                   5: "After Work Rush (16-18)",
                   6: "Evening (18-22)",
                   7: "Night (22-6)"}
```

```python
#pull time data and create hour column
df['Hour'] = df['Time'].str[0:2]
 
#convert to numeric    
df['Hour'] =  pd.to_numeric(df['Hour'])
# drop null values in our new column
df = df.dropna(subset=['Hour'])
#convert to integer
df['Hour'] = df['Hour'].astype('int')
```

```python
#create time_of_day grouping

def daygroup(hour):
    if hour >= 6 and hour < 10:
        return "1"
    elif hour >= 10 and hour < 12:
        return "2"
    elif hour >= 12 and hour < 14:
        return "3"
    elif hour >= 14 and hour < 16:
        return "4"
    elif hour >= 16 and hour < 18:
        return "5"
    elif hour >= 18 and hour < 22:
        return "6"
    else:
        return "7"
```

```python
#apply function   
#time of day function
df['Time']= df['Hour'].apply(daygroup)   
```

```python
df[['Weekend','Day_of_Week', 'Time']].tail(10)
```

```python
# number_of_vehicles
df['Number_of_Vehicles'].value_counts()/df.shape[0]*100
```

```python
#group number_of_vehicles

def vehicles(num_veh):
    if num_veh >=1 and num_veh <2:
        return "1"
    elif num_veh >=2 and num_veh <3:
        return "2"
    elif num_veh >=3 and num_veh <4:
        return "3"
    elif num_veh >= 4: 
        return "4+"
  
#apply function   
df['Number_of_Vehicles']= df['Number_of_Vehicles'].apply(vehicles)
```

```python
# number_of_vehicles
df['Number_of_Vehicles'].value_counts()/df.shape[0]*100
```

```python
df['Number_of_Vehicles'].dtypes
```

```python
df['Number_of_Vehicles']=df['Number_of_Vehicles'].astype('object')
```

```python
#creating seasons column for ML

#creating season column

def getSeason(month):
    if (month == 12 or month == 1 or month == 2):
       return "winter"
    elif(month == 3 or month == 4 or month == 5):
       return "spring"
    elif(month == 6 or month== 7 or month == 8):
       return "summer"
    else:
       return "fall"

df['Season'] = df['Month'].apply(getSeason)
```

```python
# number_of_vehicles
df['Season'].value_counts()/df.shape[0]*100
```

```python
#Put above pickle in next full run
#create new column for Machine Learning and Visualizations with Not Severe and Severe
df['accident_seriousness'] = df['Accident_Severity']
df['accident_seriousness'] = df['accident_seriousness'].replace(to_replace="Slight", 
                                                                value="Not Severe")
df['accident_seriousness'] = df['accident_seriousness'].replace(to_replace="Serious",
                                                                value="Severe")
df['accident_seriousness'] = df['accident_seriousness'].replace(to_replace="Fatal", 
                                                                value="Severe")
df.shape
```

```python
def probability_of_accident_seriousness(col):
    """Produces a table of the proportion of accidents in each category of col that were severe or not_severe"""
    return df.groupby(col)['accident_seriousness'].value_counts(normalize=True).unstack()
```

```python
probability_of_accident_seriousness('Did_Police_Officer_Attend_Scene_of_Accident')
```

```python
def unsorted_severity_category_plot(col, title='Proportion of accidents by severity', figsize=(10,5)):
    """Plots a bar chart of the proportions of the specified column for each accident severity category"""
    df.groupby('accident_seriousness')[col].value_counts(normalize=True).unstack().sort_index(ascending=False).plot(figsize=figsize, kind='bar')
    plt.title(title)
    plt.xlabel('accident_seriousness')
    plt.ylabel('Proportion of accidents')
    plt.xticks(rotation=0)
    plt.legend();
```
